```python
df['Family_size'] = df['Parch'] + df['SibSp']
```

```python
df.sample(5)
```

```python
def Family_type(number):
    if number ==0 :
        return "Alone"
    elif number >1 or number < 4:
        return "Medium"
    else:
        return "Large"
```

```python
df['Family_type'] = df['Family_size'].apply(Family_type)

df.sample(5)
```

# Handling outliers

```python
# Handling outliers in Age i.e. almost normal
df=df[df['Age']<df['Age'].mean() + 3* df['Age'].std()]
df.shape
```

```python
# Handling outliers in fare column
q1,q3=np.percentile(df['Fare'],[25,75])
print('q1 is : ', q1,"\n","q3 is : ", q3)
```

```python
IQR = q3 - q1
Lower_Fence = q1 -1.5*(IQR)
Upper_Fence = q3 +1.5*(IQR)
print("Lower_Fence is : ", Lower_Fence,"\n","Upper_Fence is : ", Upper_Fence)
```

```python
df = df[(df['Fare']>Lower_Fence) &  (df['Fare']<Upper_Fence)]
df.head(5)
```

# one hot encoding

```python
df=pd.get_dummies(data=df, columns=['Pclass','Sex','Embarked','Family_type'],
                   drop_first=True).replace([True,False],[1,0])
```

```python
df.drop(['Name','Ticket'],axis=1,inplace=True)
```
